```python
import requests
import pandas as pd
from parse_ct import *
```

```python
query="commissariat"
```

```python
nb_pages = 1
```

```python
data = []
for p in range(0, nb_pages):
    min_rnk = 1 + 100*p
    max_rnk = (p+1) * 100
    r = requests.get(f'https://clinicaltrials.gov/api/query/full_studies?expr={query}&fmt=json&min_rnk={min_rnk}&max_rnk={max_rnk}')
    data += r.json()['FullStudiesResponse']['FullStudies']
len(data)
```

```python
r.json()['FullStudiesResponse']['NStudiesFound']
```

```python
res = []
```

```python
for ix, k in enumerate(data):
    study = parse_study(k)
    if study not in res:
        res.append(parse_study(k))
```

```python
df = pd.DataFrame(res)#[['NCTId', 'title', 'study_completion_date', 
#'study_type', 'has_publications_result', 'has_results', 'has_results_or_publications', 'lead_sponsor', 'collaborators', 'sponsor_collaborators']]
```

```python
df['cea'] = False
for ix, row in df.iterrows():
    has_cea = False
    for e in row.sponsor_collaborators:
        if 'commissariat' in e.lower() or 'cea;' in (e+';').replace(' ', ';').lower():
            has_cea = True
    if has_cea is False:
        print(row.NCTId, end=',')
    else:
        df.at[ix, 'cea'] = True
```

```python
df[df.cea]
```

```python
df_cea = df[df.cea]
```
