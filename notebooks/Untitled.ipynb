```python
import pandas as pd
```

```python
df = pd.read_csv('/Users/eric/Downloads/bso-clinical-trials-2022-01-21_20220121.csv.gz')
```

```python
def get_ids(r):
    ids = []
    for k in ['eudraCT', 'NCTId']:
        if not(pd.isnull(r[k])):
            ids.append(k)
    return ' + '.join(ids)
```

```python
df['id_type'] = df.apply(lambda r:get_ids(r), axis=1)
```
